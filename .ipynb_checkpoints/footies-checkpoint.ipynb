## Transfer learning

* Instead of loading each image one-by-one, we can use a data generator ImageDataGenerator. Keras will use it for loading the images and pre-processing them
* Train Xception on smaller (150 x 150)

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_gen = ImageDataGenerator(preprocessing_function=preprocess_input)
```

```python
image_size = (150, 150)
batch_size = 32
```

```python
train_ds = train_gen.flow_from_directory(
    "zalando-dataset/train",
    seed=1,
    target_size=image_size,
    batch_size=batch_size,
)
```

```python
X, y = next(train_ds)
```

```python
X.shape
```

```python
val_gen = ImageDataGenerator(preprocessing_function=preprocess_input)
```

```python
 val_ds = val_gen.flow_from_directory(
    "zalando-dataset/validation",
    target_size=image_size,
    batch_size=batch_size,
     
)
```

```python
base_model = Xception(
    weights='imagenet',
    include_top=False,
    input_shape=(150, 150, 3)
    
)

base_model.trainable = False 
```

```python
inputs = keras.Input(shape=(150, 150, 3))

base = base_model(inputs, training=False)
vector = keras.layers.GlobalAveragePooling2D()(base)

outputs = keras.layers.Dense(4)(vector)

model = keras.Model(inputs, outputs) 
```

```python
learning_rate = 0.01


optimizer = keras.optimizers.Adam(learning_rate)

loss = keras.losses.CategoricalCrossentropy(from_logits=True)

model.compile(
    optimizer=optimizer, 
    loss=loss, 
    metrics=["accuracy"]
)
```

```python
model.fit(train_ds, epochs=10, validation_data=val_ds)
```

```python
history = model.fit(train_ds, epochs=10, validation_data=val_ds)
```

```python
plt.plot(history.history['accuracy'], label='train')
plt.plot(history.history['val_accuracy'], label='val')
plt.xticks(np.arange(10))
plt.legend()
```

### Adjusting the learning rate
* What's the learning rate
* Trying different values and see what best for the model

```python
def make_model(learning_rate):
    base_model = Xception(
    weights='imagenet',
    input_shape=(150, 150, 3),
    include_top=False
    )
    
    base_model.trainable = False 
    
    #########################################
    
    inputs = keras.Input(shape=(150, 150, 3))
    base = base_model(inputs, training=False)
    vector = keras.layers.GlobalAveragePooling2D()(base)
    outputs = keras.layers.Dense(4)(vector)
    model = keras.Model(inputs, outputs) 

    #########################################
        
    optimizer = keras.optimizers.Adam(learning_rate)
    loss = keras.losses.CategoricalCrossentropy(from_logits=True)
    model.compile(
        optimizer=optimizer, 
        loss=loss, 
        metrics=['accuracy']
    )
    
    return model
```

```python
scores = {}

for lr in [0.0001, 0.001, 0.01, 0.1]:
    print(lr)
    model = make_model(learning_rate=lr)
    history = model.fit(train_ds, epochs=10, validation_data=val_ds)
    scores[lr] = history.history
    
    print()
    print()
```

```python
preds = model.predict(X)
```

```python
del scores[0.1]
del scores[0.0001]
```

```python
for lr, hist in scores.items():
    
    #plt.plot(hist['accuracy'], label='train=%s' % lr)
    plt.plot(hist['val_accuracy'], label=('val=%s' % lr))
    
plt.xticks(np.arange(10))
plt.legend()
```
